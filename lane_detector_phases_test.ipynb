```python
from lane_detection import edge_detection, get_curve, perspective_warp, sliding_window, draw_lanes

import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_image (image, title="Image", cmap_type="gray"):
    plt.imshow(image, cmap_type)
    plt.title(title)
    plt.axis("on")
```

```python
def display_images(img_path):
    
    original_image = plt.imread("./test_images/"+img_path)
    images = []
    img = edge_detection(original_image, filter_color=True)
    images.append(img.copy())
    img = perspective_warp(img)
    images.append(img.copy())
    out_img, curves, lanes, ploty = sliding_window(img)
    images.append(out_img.copy())

    curverad = get_curve(original_image, curves[0], curves[1])
    lane_curve = np.mean([curverad[0], curverad[1]])
    img = draw_lanes(original_image, curves[0], curves[1])
    images.append(img.copy())

    fig, axs = plt.subplots(2,2, figsize=(15, 10))
    for im,ax in zip(images, axs.flat):
        ax.imshow(im, "gray")
```
